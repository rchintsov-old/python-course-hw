### Постройте таблицу в которой по годам всех олимпиад покажите топовых атлетов США(1 атлет на год) по общему количеству медалей? Включите дисциплину атлета.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
oo = pd.read_csv('Summer Olympic medallists 1896 to 2008 - ALL MEDALISTS.csv', skiprows=4)
oo.head()
```

```python
# создает таблицу, где столбцы - это атлеты, а строки - годы олимпиад
athletes_by_year = oo[oo.NOC == 'USA'].groupby(['Edition', 'Athlete', 'Medal']).size() \
.unstack('Medal', fill_value=0).sum(axis=1).unstack('Athlete', fill_value=0)

# athletes_by_year.head()
```

```python
# создает предварительную таблицу результатов (без дисциплины атлета)
df = pd.concat([athletes_by_year.max(axis=1), athletes_by_year.idxmax(axis=1)], axis=1).reset_index()

# df.head()
```

```python
# функция, которая выбирает наиболее "призовую" дисциплину для атлета в отпределенный год из основной таблицы
select_discipline = lambda x: oo[(oo.Edition == x.Edition) & (oo.Athlete == x[1])] \
.groupby(['Medal', 'Discipline']).size().unstack('Discipline').sum().idxmax()
```

```python
# добавляет дисциплину в df и переименовывает столбцы
result = df.merge(df.apply(select_discipline, axis=1).to_frame(), left_index=True, right_index=True)
result.columns = ['Edition', 'Medals (total in year)', 'Name', 'Discipline']
```
